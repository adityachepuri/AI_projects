```python
import lazypredict
```

```python
from lazypredict.Supervised import LazyClassifier
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
X=pd.read_csv('/content/drive/MyDrive/AI/steel_grade_prediction/med_proj/X.csv')
y=pd.read_csv('/content/drive/MyDrive/AI/steel_grade_prediction/med_proj/y.csv')
```

```python
y.head()
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=45)
```

```python
clf=LazyClassifier(verbose=0,ignore_warnings=True,custom_metric=None)
```

```python
models,predictions=clf.fit(X_train,X_test,y_train,y_test)
print(models)
```

```python
predictions
```

```python
predictions.to_excel('pred.xls')
```

```python
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
plt.figure(figsize=(10,5))
sn.set_theme(style='whitegrid')
ax=sn.barplot(x=models.index,y='Accuracy',data=models)
plt.xticks(rotation=90)
```

Bagging Classifier, XGBoost Classifier, Decision Tree, Linear Discriminant Analysis and RandomForest Classifiers performed better compared to other.
