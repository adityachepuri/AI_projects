```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
```

**Step 1:** Import InceptionV3 application

```python
model=tf.keras.applications.InceptionV3(weights='imagenet')
model.summary()
```

**Step 2:** Load Image

```python
image=tf.keras.preprocessing.image.load_img('/content/drive/MyDrive/AI/adv_dl_computer_vision/data_sets/simplilearn_dataset/elephant.jpg',target_size=(299,299))
plt.imshow(image)
```

**Step 3:** Convert Image into numpy array

```python
image_to_array=tf.keras.preprocessing.image.img_to_array(image)
image_to_array.shape
```

**Step 4:** Making image_to_array compatible by adding dimension n

```python
image_compatible=np.expand_dims(image_to_array,axis=0)
image_compatible.shape
```

**Step 5:** Making predictions

```python
preds=model.predict(image_compatible)
preds
```

```python
tf.keras.applications.InceptionV3.decode_predictions(preds,top=5)
```
