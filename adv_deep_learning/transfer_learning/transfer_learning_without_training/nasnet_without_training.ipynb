```python
import tensorflow as tf 
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
```

**Step 1:** Import Model

```python
model=tf.keras.applications.NASNetLarge(weights='imagenet')
model.summary
```

**Step 1:** Import image

```python
image=tf.keras.preprocessing.image.load_img('/content/drive/MyDrive/AI/adv_dl_computer_vision/data_sets/simplilearn_dataset/elephant.jpg',target_size=(331,331))
plt.imshow(image)
```

**Step 2:**Convert imput image into array

```python
image_to_array=tf.keras.preprocessing.image.img_to_array(image)
image_to_array.shape
```

**Step 3:** Convert image_to_array compatible by adding n 

```python
image_to_compatible=np.expand_dims(image_to_array,axis=0)
image_to_compatible.shape
```

**Step 4:** Prediction

```python
preds=model.predict(image_to_compatible)
```

**Step 5:** Decode the predictions

```python
tf.keras.applications.nasnet.decode_predictions(preds,top=5)[0]
```
