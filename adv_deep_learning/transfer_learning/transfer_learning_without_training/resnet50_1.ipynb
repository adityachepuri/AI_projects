```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
model = tf.keras.applications.ResNet50(weights='imagenet')
```

```python
model.summary()
```

```python
image_to_predict=tf.keras.preprocessing.image.load_img('/content/drive/MyDrive/AI/adv_dl_computer_vision/data_sets/simplilearn_dataset/elephant.jpg',target_size=(224,224))
```

```python
plt.imshow(image_to_predict)
```

```python
image_to_array=tf.keras.preprocessing.image.img_to_array(image_to_predict)
```

```python
image_to_array.shape
```

```python
# Step 3: Make the image compatible for the model by adding dimensions
my_compatible=np.expand_dims(image_to_array, axis=0)
my_compatible.shape
```

```python
# Step4:
prediction=model.predict(my_compatible)
prediction
```

```python
# Step 5: 
tf.keras.applications.resnet50.decode_predictions(prediction,top=5)
```
