```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
import numpy as np
```

**Step 1:** Load Model

```python
model=tf.keras.applications.InceptionV3(include_top=True, weights='imagenet')
```

```python
model.summary()
```

**Step 2:** Load the image

```python
image=tf.keras.preprocessing.image.load_img('/content/drive/MyDrive/AI/adv_dl_computer_vision/data_sets/simplilearn_dataset/basket_ball_1.jpg',target_size=(299,299))
```

```python
plt.imshow(image)
```

**Step 3:** Convert image to array

```python
image_to_array=tf.keras.preprocessing.image.img_to_array(image)
```

```python
image_to_array.shape
```

**Step 4:** Make the image comaptible for model by adding n

```python
image_compatible=np.expand_dims(image_to_array,axis=0)
```

**Step 4:** Prediction

```python
pred=model.predict(image_compatible)
```

**Step 6:** Decode Prediction

```python
tf.keras.applications.inception_v3.decode_predictions(pred,top=5)
```
