```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pyplot as plt
```

```python
tf.__version__
```

```python
model=tf.keras.applications.VGG16(weights='imagenet')
```

```python
model.summary()
```

**Step 1:** Loading Image

```python
image=tf.keras.preprocessing.image.load_img('/content/drive/MyDrive/AI/adv_dl_computer_vision/data_sets/simplilearn_dataset/basket_ball_1.jpg',target_size=(224,224))
```

```python
plt.imshow(image)
```

**Step 2 :** Convert Image to Array

```python
image_to_array=tf.keras.preprocessing.image.img_to_array(image)
```

```python
image_to_array.shape
```

**Step 3:** Make the image compatible for the model by adding n

```python
image_compatible=np.expand_dims(image_to_array,axis=0)
image_compatible.shape
```

**Step 4:** Prediction

```python
pred=model.predict(image_compatible)
pred
```

**Step5 :** Decoding Prediction

```python
tf.keras.applications.vgg16.decode_predictions(pred, top=5)
```
